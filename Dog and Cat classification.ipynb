```python
input_dir = r"C:\Users\Ishwari\Downloads\test_set\test_set"
categories = ['cats', 'dogs']
data = []
labels = []
for category_idx, category in enumerate(categories):
    for file in os.listdir(os.path.join(input_dir, category)):
        img_path = os.path.join(input_dir, category, file)
        print(img_path)
        img = imread(img_path)
        img = resize(img, (15,15))
        data.append(img.flatten())
        labels.append(category_idx)    
```

```python
data[1]
```

```python
labels[1]
```

```python
data = np.asarray(data)
labels = np.asarray(labels)
```

```python
data
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels)
```

# Built SVM Model for classification

```python
from sklearn.svm import SVC
svm_model = SVC()
```

### Model Training

```python
svm_model.fit(x_train, y_train)
```

```python
y_pred = svm_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
score
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score = cross_val_score(svm_model, data, labels, cv = 5)
cross_val_score
```

```python
Mean_Accuracy = cross_val_score.mean()
Mean_Accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifer = SVC()
parameters = [{'gamma':[0.01, 0.001, 0.0001], 'C':[10, 100, 1000]}]
grid_search = GridSearchCV(classifer, parameters)
grid_search.fit(x_train, y_train)
```

```python
best_estimator = grid_search.best_estimator_
best_estimator
```

```python
y_prediction = best_estimator.predict(x_test)
y_prediction
```
